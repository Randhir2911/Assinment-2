(Q.1) - What are data structures, and why are they important

#Answer -  Data structures are ways to organize and store data in a computer so it can be accessed and updated efficiently. They are used in almost                every program and software system. 

Why are data structures important?
1- Efficient data management: Data structures help manage large amounts of data efficiently. 
2 -Faster programs: Understanding data structures can help you create programs that run faster. 
3- Memory optimization: Data structures can help you create programs that use less memory. 
4- Fundamental to computer science: Data structures are a core part of computer science education. 

Examples of data structures
1 - Arrays-:
A fundamental data structure that stores items of the same type in adjoining memory locations 
2 - Hash tables -:
A data structure that stores key/value pairs using a hash function to compute an index into an array 
3 - Linear data structures -:
A data structure that arranges data in a line, making it easy to access and traverse elements in order. Examples include arrays, linked lists, and queues .

Cannot be changed after creation
Values or properties cannot be modified directly

(Q.3)  what are the main differences between lists and tuples in python

Answer -: In Python, lists and tuples are both data structures that store collections of data. The main differences between the two are: 

1- Mutability: Lists can be changed, while tuples cannot. 
2- Syntax: Lists use square brackets, while tuples use round brackets. 
3- Performance: Tuples are faster and use less memory than lists. 
4- Built-in methods: Lists have more built-in methods than tuples. 
5- Appropriate usage: Lists are better for data that changes frequently, while tuples are better for data that should remain the same.
